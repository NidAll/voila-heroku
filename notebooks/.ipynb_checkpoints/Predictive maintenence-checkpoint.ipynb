## Normalisation des données

```python
def z_score(df):
    # copy the dataframe
    df_std = df.copy()
    # apply the z-score method
    for column in df_std.columns:
        df_std[column] = (df_std[column] - df_std[column].mean()) / df_std[column].std()
        
    return df_std
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
```

```python
def elbow(data,steps=20):
    n_cluster = range(1, steps)
    kmeans = [KMeans(n_clusters=i).fit(data) for i in n_cluster]
    scores = [kmeans[i].score(data) for i in range(len(kmeans))]
    
    df = pd.DataFrame(dict(
        n_cluster = n_cluster,
        scores = scores
    ))
    
    fig =px.line(df, x=df['n_cluster'], y=df['scores'], title="Elbow Curve")
    #plt.xlabel('Number of Clusters')
    #plt.ylabel('Score')
    #plt.title('Elbow Curve')
    fig.show();
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_781K001, y_781K001, test_size=0.25, random_state=42)
```

```python
(unique_train, counts_train) = np.unique(y_train, return_counts=True)
(unique_test, counts_test) = np.unique(y_test, return_counts=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree_model_781K001 = DecisionTreeClassifier(max_depth = 4).fit(X_train, y_train)
dtree_predictions = dtree_model_781K001.predict(X_test)
```

```python
cm = confusion_matrix(y_test, dtree_predictions)
```

```python
#print(dtree_predictions)
#print(y_test)
#np.array_equiv(dtree_predictions,y_test)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
max_depths = np.linspace(1, 32, 32, endpoint=True)
F1_scores = []
precisions = []
recalls = []
accuracies = []
for max_depth in max_depths:
   dt = DecisionTreeClassifier(max_depth=max_depth)
   dt.fit(X_train, y_train)
   y_pred = dt.predict(X_test)
   recall = recall_score(y_test, y_pred,average='micro')
   precision= precision_score(y_test, y_pred,average='micro')
   accuracy = accuracy_score(y_test, y_pred)
   
   accuracies.append(accuracy)
   recalls.append(recall)
   precisions.append(precision)
```

```python
#accuracies
```

```python
#recalls
```

#### Optimisation des hyperparamètres

```python
df = pd.DataFrame(dict(
    tree_depth = np.linspace(1, 32, 32, endpoint=True),
    accuracies = accuracies
))
    
fig =px.line(df, x=df['tree_depth'], y=df['accuracies'])
fig.show();
```
